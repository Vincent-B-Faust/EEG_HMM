# EEG_HMM Interactive Workflow

这个 notebook 使用与附件 `analyze_emg_signal.ipynb` 一致的参数命名：`filename`、`fs`、`use_dask`、`k_user`、`window_size`、`overlap`。

```python
from sleep_hmm import run_interactive

filename = "path/to/data.edf"
fs = 2000
use_dask = True
k_user = 3
window_size = None
overlap = None

eeg_channel = "EEG"
emg_channel = "EMG"
mat_variable = None
csv_eeg_column = None
csv_emg_column = None
csv_has_header = None
csv_separator = None

feature_mode = "full"
feature_scaling = "zscore"
window_strategy = "notebook_auto"
acceleration_backend = "auto"
acceleration_device = "auto"
output_dir = "notebook_outputs"
```

```python
result = run_interactive(
    filename=filename,
    fs=fs,
    use_dask=use_dask,
    k_user=k_user,
    window_size=window_size,
    overlap=overlap,
    output_dir=output_dir,
    feature_mode=feature_mode,
    feature_scaling=feature_scaling,
    window_strategy=window_strategy,
    acceleration_backend=acceleration_backend,
    acceleration_device=acceleration_device,
    eeg_channel=eeg_channel,
    emg_channel=emg_channel,
    mat_variable=mat_variable,
    csv_eeg_column=csv_eeg_column,
    csv_emg_column=csv_emg_column,
    csv_has_header=csv_has_header,
    csv_separator=csv_separator,
)

result.output_dir
```
